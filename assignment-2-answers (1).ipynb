```python
#Q6.List all categorical and numerical features separately.
##ANSWER:-
categorical_features = df.select_dtypes(include=['object']).columns
numerical_features = df.select_dtypes(include=['int64', 'float64']).columns
print("Categorical Features:")
print(categorical_features)
print("\nNumerical Features:")
print(numerical_features)
```

```python
#7.Plot the distribution of customer ages using a histogram.
##ANSWER:-
plt.hist(df['Age'], bins=20, color='purple', edgecolor='black')
plt.title('Distribution of Customer Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
```

```python
#Q8.Create a count plot showing the number of customers in each occupation.
##ANSWER:-
sns.countplot(data=df, x='Occupation',hue='Occupation', palette='Set1',legend=False)
plt.title('Number of Customers in Each Occupation')
plt.show()
```

```python
#Q9.Plot the distribution of monthly income categories.
##ANSWER:-
sns.countplot(x='Monthly Income',hue='Monthly Income',data=df,palette="Set3",legend=False)
plt.title('Distribution of Monthly Income Categories')
plt.xlabel('Monthly Income')
plt.ylabel('Number of Customers')
plt.xticks(rotation=10)
plt.show()
```

```python
#Q10.Find the top 5 most common educational qualifications.
##ANSWER:-
df['Educational Qualifications'].value_counts().head(5)
```

```python
#Q11.Compare the distribution of Gender across different Occupations using an appropriate visualization.
##ANSWER:-
sns.countplot(data=df, x='Occupation',hue='Gender', palette='Set2')
plt.title('Distribution of Gender Across Different Occupations')
plt.show()
```

```python
#Q12.Analyze the relationship between Marital Status and Feedback.
##ANSWER:-
sns.countplot(data=df, x='Marital Status',hue='Feedback',palette='dark:red')
plt.title('Relationship between Marital Status and Feedback')
plt.show()

```

```python
#Q13.Compare Family Size across different Monthly Income groups using a box plot
##ANSWER:-
sns.boxplot(data=df,x='Monthly Income',hue='Monthly Income',y='Family size',palette='pastel',legend=False)
plt.title('Compare Family Size across different Monthly Income groups')
plt.xticks(rotation=20)
plt.show()
```

```python
#Q14.Visualize the relationship between Age and Family Size using a scatter plot.
##ANSWER:-
sns.scatterplot(data=df,x='Age',y='Family size',color='red')
plt.title('Relationship between Age and Family Size')
plt.show()
```

```python
#Q15.Create a heatmap of correlations among all numerical features. 
##AANSWER:-
numerical_features = df.select_dtypes(include=['int64', 'float64'])
correlation_matrix = numerical_features.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
#Q16.Analyze how Occupation, Monthly Income, and Feedback are related. 
##ANSWER:-
sns.barplot(data=df,x='Occupation',y='Monthly Income',hue='Feedback')
plt.title('Relationship between Occupation, Monthly Income, and Feedback')
plt.xticks(rotation=20)
plt.show()
```

```python
#Q17.Compare the average family size for each educational qualification.
##ANSWER:-
df.groupby('Educational Qualifications')['Family size'].mean()
```

```python
#Q18.Plot customer locations using Latitude and Longitude on a scatter plot and identify any visible clusters.
##ANSWER:-
sns.scatterplot(data=df,x='latitude',y='longitude',color='blue')
plt.title('Customer Locations')
plt.show()
```

```python
#Q19.Determine which Pin Codes have the highest number of customers.
##ANSWER:-
df.groupby('Pin code').size().sort_values(ascending=False).head(1)
```

```python
#Q20.Perform a complete EDA and summarize at least five meaningful business insights
#that could help an online food delivery company improve its services.

# ============================
#  Complete EDA Summary
# ============================

print("Dataset Shape:")
print(df.shape)

print("\nDataset Info:")
print(df.info())

print("\nMissing Values:")
print(df.isnull().sum())

print("\nSummary Statistics:")
print(df.describe())

print("\nCategorical Features Summary:")
print(df.describe(include='object'))

# Top Occupations
print("\nTop Occupations:")
print(df["Occupation"].value_counts())

# Monthly Income Distribution
print("\nMonthly Income Distribution:")
print(df["Monthly Income"].value_counts())

# Feedback Distribution
print("\nFeedback Distribution:")
print(df["Feedback"].value_counts())

# Top Pin Codes
print("\nTop Pin Codes by Customer Count:")
print(df["Pin code"].value_counts())

# Correlation between numerical columns
print("\nCorrelation Matrix:")
print(df.corr(numeric_only=True))

##HERE IS SOME BUSSINESS INSIGHTS BY USING ONLINE FOOD DATASET


#1]The majority of customers belong to the most common occupation.
#2]Most customers fall into a specific monthly income category.
#3]Positive feedback is higher than negative feedback, indicating good customer satisfaction.
#4]A few Pin Codes contribute the highest number of customers, showing high-demand areas.
#5]Numerical features with strong correlations can help the company understand customer behavior.
```
